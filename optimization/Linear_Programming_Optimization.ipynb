```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# Define the coefficients of the objective function
c = [-3, -4]  # Minimize -3x - 4y (since linprog does minimization by default)

# Define the inequality constraints matrix
A = [
    [2, 3],  # 2x + 3y <= 24 (Labor hours)
    [1, 2],  # x + 2y <= 6 (Machine hours)
]

# Define the inequality constraints vector
b = [24, 6]

# Define the bounds for x and y
x_bounds = (0, None)
y_bounds = (0, None)

# Solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Extract the solution
x_opt, y_opt = result.x
z_opt = -result.fun

# Define the constraints
x = np.linspace(0, 20, 400)
y1 = (24 - 2*x) / 3   # 2x + 3y <= 24
y2 = (6 - x) / 2      # x + 2y <= 6

# Create a plot
plt.figure(figsize=(10, 8))

# Plot the constraint lines
plt.plot(x, y1, label=r'$2x + 3y \leq 24$', color='red')
plt.plot(x, y2, label=r'$x + 2y \leq 6$', color='blue')

# Fill the feasible region
y_max = np.minimum(y1, y2)
plt.fill_between(x, 0, y_max, where=(y_max >= 0), color='lightgreen', alpha=0.5)

# Plot the optimal solution
plt.plot(x_opt, y_opt, 'go', label='Optimal solution')
plt.text(x_opt, y_opt, f'  ({x_opt:.2f}, {y_opt:.2f})', verticalalignment='bottom', horizontalalignment='right')

# Plot the axes and labels
plt.xlim(0, 20)
plt.ylim(0, 10)
plt.xlabel('$x$ (units of product A)')
plt.ylabel('$y$ (units of product B)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()

# Title
plt.title('Feasible Region and Optimal Solution for the Linear Programming Problem')

# Show the plot
plt.show()

print(f'Optimal solution: x = {x_opt}, y = {y_opt}')
print(f'Maximized profit: Z = ${z_opt}')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import linprog

# Define the coefficients of the objective function
obj = [-1, -2]

# Coefficients for the inequality constraints
lhs_ineq = [[2, 1],   # Red constraint left side
            [-4, 5],  # Blue constraint left side
            [1, -2]]  # Yellow constraint left side

# Constants for the inequality constraints
rhs_ineq = [20,  # Red constraint right side
            10,  # Blue constraint right side
            2]   # Yellow constraint right side

# Coefficients for the equality constraint
lhs_eq = [[-1, 5]]  # Green constraint left side

# Constants for the equality constraint
rhs_eq = [15]       # Green constraint right side

# Bounds for the variables
bnd = [(0, float("inf")),  # Bounds of x
       (0, float("inf"))]  # Bounds of y

# Perform the linear programming optimization
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,
              A_eq=lhs_eq, b_eq=rhs_eq, bounds=bnd,
              method="revised simplex")


# Print the results
print("Optimal value:", opt.fun)
print("Success:", opt.success)
print("Optimal solution:", opt.x)

# Plot the constraints
x = np.linspace(0, 10, 500)

# Red constraint: 2x + y <= 20
y1 = (20 - 2*x)

# Blue constraint: -4x + 5y <= 10
y2 = (10 + 4*x) / 5

# Yellow constraint: x - 2y <= 2
y3 = (x - 2) / 2

# Green constraint: -x + 5y = 15
y4 = (15 + x) / 5

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r'$2x + y \leq 20$ (Red)', color='r')
plt.plot(x, y2, label=r'$-4x + 5y \leq 10$ (Blue)', color='b')
plt.plot(x, y3, label=r'$x - 2y \leq 2$ (Yellow)', color='y')
plt.plot(x, y4, label=r'$-x + 5y = 15$ (Green)', color='g')

# Fill feasible region
plt.fill_between(x, np.maximum(y1, y3), np.minimum(y2, y4), where=(y1 <= y4) & (y1 <= y2), color='grey', alpha=0.5)

# Mark the optimal solution
plt.plot(opt.x[0], opt.x[1], 'ro', label='Optimal Solution')
plt.annotate(f'Optimal Solution\nx={opt.x[0]:.2f}, y={opt.x[1]:.2f}',
             xy=(opt.x[0], opt.x[1]), xytext=(opt.x[0] + 1, opt.x[1] - 1),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Feasible Region and Optimal Solution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
model = LpProblem(name="resource-allocation", sense=LpMaximize)

# Define the decision variables
x = {i: LpVariable(name=f"x{i}", lowBound=0) for i in range(1, 5)}
y = {i: LpVariable(name=f"y{i}", cat="Binary") for i in (1, 3)}

# Add constraints
model += (lpSum(x.values()) <= 50, "manpower")
model += (3 * x[1] + 2 * x[2] + x[3] <= 100, "material_a")
model += (x[2] + 2 * x[3] + 3 * x[4] <= 90, "material_b")

M = 100
model += (x[1] <= y[1] * M, "x1_constraint")
model += (x[3] <= y[3] * M, "x3_constraint")
model += (y[1] + y[3] <= 1, "y_constraint")

# Set objective
model += 20 * x[1] + 12 * x[2] + 40 * x[3] + 25 * x[4]

# Solve the optimization problem
status = model.solve()

print(f"status: {model.status}, {LpStatus[model.status]}")
print(f"objective: {model.objective.value()}")

for var in model.variables():
    print(f"{var.name}: {var.value()}")

for name, constraint in model.constraints.items():
    print(f"{name}: {constraint.value()}")
```
